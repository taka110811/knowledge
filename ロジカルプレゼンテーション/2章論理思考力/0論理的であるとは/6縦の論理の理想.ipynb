論理では、「求められれば細かくできる」状態に自分を鍛えておくことが大事である。

（１）の「考えていない人」は、路上駐車は迷惑だという言葉をそのまま鵜呑みにしており、それ以上何も考えていない人を指す。\
（２）の「考えすぎの人」は、路上駐車はなぜ迷惑なのかということをくどくどと考え、多くの理屈をこねている人を指す。\
（３）の「求められれば細かくできる人」は、路上駐車は迷惑だというのを論理的に理解したうえで、説明を求められれば細かくもできるという人である。\


（１）のタイプは、いわゆる直感型と言われる人に多い考え方である。このタイプの人の場合、直感が正しい範囲ではなんら問題はないが、直感がまちがっていた場合は手の打ちようがない。「路上駐車は迷惑よ」「なんで？」「迷惑なものは迷惑」といった具合で、会話にならないからだ。\
（２）のタイプは、つねに理屈をこねている人に多い考え方で、これだと話が長くて時間がかかる。\
理想は（３）である。自分が（３）のタイプになれれば（１）（２）（３）すべてのタイプに話を理解させることができる。\
（１）のタイプの人と話をするときは、飛躍感のある論理で相手の直感にあわせて話をする。\
（２）のタイプの人と話をする場合は、事細かく話す。\
（３）同士で話をするときは、必要箇所だけに絞って手短に話す。
