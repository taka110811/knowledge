横の論理では、「できるだけ広く細かい」状態が作り出せるように自分の頭を鍛えておくことが大事である。

この場合も、目指すは（３）である。（１）や（２）のタイプだと、いつまでたっても議論はかみあわない。\
「細かいことしか分からない部下と、細かいことが分からない上司」「大きな話しかできない上司と、大きな話が分からない部下」のような関係である。\
ここでも自分が（３）のタイプになれれば、すべてのタイプに話を理解させることができる。 　

漏れを議論するうえでの「全体」のとらえ方は相手によってちがうと述べたが、\
相手がどんなタイプであっても、自分の考えを理解させるためには、つねに「できるだけ広く細かい論理が組める」ように自分を鍛えておかなければならない。
