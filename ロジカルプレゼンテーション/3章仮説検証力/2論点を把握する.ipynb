### 論点を噛み合わせるための方法


### 会議や提案の場で「気の利いた発言」ができる人
「どこが原因で議論が停滞しているのか」を瞬時にして実に的確に捉えている<br>
論点を外さないためには、論理的思考とアナログ的思考の双方のバランスを常に意識していく必要がある。

例.
* フレームワーク的な視点が欠けている会議では議論の視点について助け舟を出す
* 視点の違う意見を出す人がいても、その人の味方を変えて議論のフレームワークの中に収める
* 他の参加者の誰もが知っている当たり前の内容について話が出た場合には、上手く話を先に進める
