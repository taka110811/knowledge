# 仮説とは何か

「論点に対するヤマカンの答え」である<br>


### 仮説についての３つの誤解
1. 論点のないところに仮説はない：論点になっていない答えは仮説ではない
2. 仮説は全くの当てずっぽうではない：「仮の答え」であり、確信はなくても自分の感覚や経験、常識などに基づくもの
3. 仮説は答えではない：仮説と答えの決定的な違いは、客観的根拠に基づく裏づけがあるかないかである。「仮説と答えの混同」はビジネスの場面で多く見られる。感覚や経験は人と共有できないため客観的な事実とならない。

### なぜ仮説が必要なのか
論点がわかっていても、応えを持ち合わせていない場合の選択肢は２つしかない。<br>
事前情報を元にある程度絞ってから仮説を提示することで、時間効率よく提案することができる。

1. 考えうる全ての選択肢を相手に提示して、選んでもらう（絨毯爆発型）
2. 自分がヤマカンで選んでから、相手に提示して、選んでもらう（仮説検証型）

### 仮説検証型は相手に考えさせない
絨毯爆発型は、あくまでも考えられる全ての選択肢を提示しているだけなので、悪く言えば頭を使っていない。そのかわり相手は膨大な選択肢の中から全てを吟味する必要がある。<br>
対して仮説検証型は、こちらは必死で相手の思惑なども探りながら、いろいろ考えてヤマカンを精度を高めようと努力する。相手側はこちらのヤマカンが当たっていれば、承認すればよいだけだ。

つまり目上の人や顧客に対して話すとき、「考える労力」は自分で持つべきなため、仮説検証型の思考が必要になる。

重要なことは
自分で可能な限り考えた上で、「ABCの選択肢がありますが、こういう理由からAが良いと思います。こちらでいかがでしょうか」という形で仮説を持った質問をすることである。
