## 目的を理解する

コミュニケーションの目的を理解するポイント２つ
* 議論のスタンス
* 相手の要望

### 議論のスタンス：ビジネス上のコミュニケーションで話しかける時の期待していること
* 何か意思判断をしてもらいたい：
* 単に聞いてもらいたいだけ

ビジネスの現場ではスタンスの違いの認識不足が原因で意思疎通ができない場面が多く存在する

### 「押し切る」「引き出す」

議論のスタンスを上手く使うことで、「相手に行動を起こさせる」という結果を引き出せる。

「何か意思判断を求める」は「押し切る」コミュニケーションに分類される。\
「ただ単に聞いてもらうだけ」は「引き出す」コミュニケーションに分類される。

ときと場合に使い分けるべきだが、通常人は得意な方のみを使いがち

本書では「押し切る」コミュニケーションについてのみ触れる。「引き出す」コミュニケーションについてはコーチングやインタビューの本を読むとよい\
理由としては「押し切る」コミュニケーションが必要な場面が圧倒的に多いため。

ビジネス上のコミュニケーションでは、自分が相手に「意思判断を求めている」ことを常に念頭に置いて、相応の論理構成で話を展開しないとならない。

### 「提案は具体的な話で締めくくる」という原則

## 相手の要望の理解

これは極めて感覚的なスキル、「相手の顔色やその場の雰囲気、過去の経験、会話の流れなどから、なんとなく推察していく」スキルである\
これを理解しないと、極めて「論理思考」偏重になってしまい、アナログスキルを無視することになってしまう。屁理屈ばかりを鍛えている人間になってしまう。（屁理屈：論理的ではあるが「論点」「仮説検証」が抜けている回答）

### アナログ的なスキルを強化するためには

「相手の話を聞くこと」しかない。「提案をする」場合、人は自分が話すことばかりを考えてしまうので、相手の話をきちんと聞く姿勢が取れなくなってしまう\
ただ、まずは「しっかりと相手の話を聞いて理解をする」ことのほうが、提案を通していくうえでは重要である。\
会議にかける資料を作成するためには、その会議の出席者がどんな人なのか、何を考えているのか、どういう反応をするの等を事前に聞き出していかないと、
相手の心に響く資料にならない。

常に相手あってのコミュニケーションであるから、「相手にどのように思われているか」を考え続けることが重要である。
> 常日頃から「気配りを続ける」だけでも、相手の要求への理解、ひいては提案力そのもに大きな違いが出ることを肝に銘じておくこと。
