### 仮説をどのようにして作るのか

仮説とは「相手の質問に対する自分なりのヤマカンの答え」である<br>
下図の矢印が仮説を作ることを示している。

* 過去の経験を元に仮説を仮説を出す：ベテランの人も「過去に蓄積された膨大な情報に基づいて判断しており、たまたま持ち合わせている情報で、当たりそうな答えを判断しているに過ぎない」
* 手近情報を可能な限り集めて答えを推測する：過去の経験がない場合はこちらをやるしかない

では、仮説出しのために情報はどのように集めれば良いか

ⅠとIIを分けることは難しいが、この違いをしっかり認識しておくことは重要である。なぜなら、「仮説を出すため」「仮説を検証するため」では求められる情報が全然違ってくるためである。

### 仮説を出すための情報
日頃から集めている広くて浅い、入手しやすい情報である

### 仮説を検証するために情報
特別に集める、狭くて深い、手の込んだ情報

上記を混同してしまうと、仮説がないのに無闇に具体的で深い調査を行なってしまったり、具体的なで精度の高い答えを求められているのに、上部だけのいい加減な答えを出してしまったりする。<br>今どちらの目的で作業をしているのかをきちんと意識するべきである。

### 仮説構築のための３つのステップ
* 論点をしっかり頭に入れる
* 常に答えは何かを意識する　WHATかHOWのオープンクエッションで考える
* とにかく多くの情報を眺める　ストック情報と呼ばれる広くて浅い情報

### 仮説を広げて、精度を高める
仮説検証力で大切かつ難しいのは「精度の高い仮説をたくさん出す」ということである。<br>
どんな人でも上記３ステップに沿えば仮説は一つは考えつく<br>
ただ難しいのは、この仮説の精度を高めることと、仮説の数を増やすことである。

ではどうすれば仮説の「当たる数」を増やすことができるか
