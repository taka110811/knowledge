いくら理屈をこね回して論理的に説明しても、相手が納得しない場合がある。「本当にそうなの？」「それだけなの？」は確実に封じ込めたはずなのに、なぜ相手は納得しないのか。

### それは相手の疑問に答えていないから

ビジネスの世界で大事なのは、相手を納得させること、腹の底から理解してもらうこと、そして何か結果につながる行動を起こしてもらうことだ。\
「論理思考」そのものは、手段であって目的ではない。「論理思考」はあくまで基本的な能力で、それだけですべてが解決できると思ったら大まちがいだ。\
理屈をこねること、正論を述べることが自己目的化してはならない。要求されていない答えをいくら必死に説明しても、相手の疑問に答えないかぎり、相手はとうてい納得しない。

では、相手の疑問をどうやって特定すればいいのだろうか。

相手の疑問に答えるという作業は２つのプロセスから成り立っている
1. まず相手の疑問を知る：論点を出す
2. 次にその疑問に対して答える：仮説を検証する

「仮説検証」とは、「まずは相手の疑問すなわち論点を洗い出し、仮の答えを推測したうえで、それに対して答えるための客観的な証拠を準備すること」だと私は理解している。

仮説検証ができていないパターン
1. 絨毯爆撃
2. 根拠なき断言

絨毯爆撃

これは１の「論点を出す部分」ができていないパターンである。この場合、相手がどこに疑問を持っているか、議論のポイントが何なのか分かっていないので、とにかくシラミ潰しに考えられる質問項目すべてに答えていくしかない。\
たしかに「相手の疑問に答える」という意味では、あらゆる作業をこなし、すべての質問に対する答えを準備するというやり方もなくはない。\
しかし、このやり方だと本来しなくてもよい無駄な作業をたくさん行うため、明らかに非効率的である。また相手から見ても、知りたくもない情報が山のように盛り込まれており、かなり理解しづらいはずだ。\
「会議の資料を適当に作ってくれ」「セールスの資料を何か作ってくれ」などといわれたことはないだろうか。\
適当に作った資料が会議の出席者や顧客の心に響けば誰も苦労しない。\
心に響く内容にしようとすると、それこそ「絨毯爆撃的」に考えられるすべての内容を資料にしなければならず、膨大な手間がかかる。\
このように、仮説を立てることができなければ、いくら時間があっても足りないという状況に陥ってしまうだろう。


根拠なき自信

これは２の「仮説を検証する部分」ができていないパターンである。この場合、相手の疑問を探りあて、あてずっぽうで一応答えてはいる。しかし、客観的な証拠が伴っていないので、その答えが本当に正しいかどうかは誰にも分からないし、説得力もない。これでは、疑問に答えたことにはならない。相手の疑問に的確に答えるためには、「仮説検証」がきちんとできることが必要なのである。
