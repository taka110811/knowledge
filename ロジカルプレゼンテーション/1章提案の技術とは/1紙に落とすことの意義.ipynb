### 紙に落とすことの意義

自分の言いたいことを紙にまとめようとしたとたんに、「何が言いたかったのだろうか」「うまく表現できない……」と手が止まってしまった経験をお持ちの方も多いのではないだろうか。\
あるいは、議事録や報告書などを作成していて、\
「耳で聞いていたときは分かったと思ったのに、いざまとめを書いてみると、ぜんぜん分かっていなかった」\
「いろいろ良い話を聞いたのに、うまくまとめられない」\
と気がついた経験をお持ちの方も多いことだろう。

不思議なもので、話で聞くとだいたいのことは理解できる、または理解できたような気分になるのだが、きっちりと紙にまとまらない話は、概して矛盾していたり、曖昧で整理できていなかったりする。

言いたいことを紙にまとめるという作業は、単に資料を作成するという意味ではなく、じつは自分の頭のなかを整理し、矛盾を解決し、的確に物事を考えているということなのだ。うまく紙に落とせるまで、考えに考え抜くことによって、自らの頭は鍛えられる。

### 紙に落とすとうまく説明できる

人前で言いたいことをうまく説明したりアピールしたりできないと感じている人も多いかもしれない。\
説明能力は通常、「プレゼンテーションの能力」や「コミュニケーションの能力」だととらえられがちだが、じつはちがう。

「うまく話せない」のは、「そもそも何を話すべきかが分かっていない」ことが原因

これに気づいてからは、言いたいことを念入りに資料としてまとめ上げることに注力した。\
その結果、今までの苦労がうそのように、すらすらとプレゼンができるようになった。\
紙に落ちない主張は、いかに話術に長けていても、うまく説明できないことを心にとめておいていただきたい。
