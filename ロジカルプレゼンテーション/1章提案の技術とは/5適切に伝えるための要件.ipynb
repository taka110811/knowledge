前提となるスキルは、議論をまとめるための会議設計力である。相手に話を伝える場合、簡単な話ならば一度ですべてを相手に伝えることができるが、こみいった話であれば何度かに分けて話をしなければならない。「最終的に言いたいこと」と「今日の会議で言いたいこと」はちがう。

　（１）メッセージ化………………最も伝えたいことを、まずは正確な日本語の文章で書く。 　（２）チャート化…………………言いたいことが一目で分かるように、図で表す。 　（３）スライド化…………………複数の図形や文章を集めて、一枚の紙にまとめる。 　（４）パッケージ化………………複数のスライドを並べて、ストーリーを持たせる。 　（５）マテリアル（資料）化……複数のパッケージを組み合わせて、その会議に最適な資料としてまとめあげる。
